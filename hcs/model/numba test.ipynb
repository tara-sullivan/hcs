```python
import numpy as np
import matplotlib.pyplot as plt
import timeit

%matplotlib inline
```

```python
=4.0
def qm(x0, n):
    x=np.empty(n)
    x[0] = x0
    for i in range(1,n):
        x[i] = *x[i-1]*(1 - x[i-1])
    return(x)

x = qm(.1,250)

fig, ax = plt.subplots()
ax.plot(x)
plt.show()
```

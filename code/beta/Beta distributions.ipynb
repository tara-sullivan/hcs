```python
import pandas as pd
import numpy as np

from scipy.stats import beta

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
a,b = 0.5, 0.5
x = np.linspace(beta.ppf(0.01,a,b),beta.ppf(0.99,a,b), 100)
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, beta.pdf(x,a,b))

ax.set_ylim(0,2.5)
```

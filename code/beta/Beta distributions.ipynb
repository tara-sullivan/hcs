```python
import pandas as pd
import numpy as np
import math

from scipy.stats import beta

import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
```

```python
a,b = 5, 5
x = np.linspace(beta.ppf(0.01,a,b),beta.ppf(0.99,a,b), 100)
```

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, beta.pdf(x,a,b))

ax.set_ylim(0,2.5)
```

```python
mu = 0.5

mu_M, var_M = mu, 0.04
mu_F, var_F = mu, 0.02

nu_M = (mu_M*(1 - mu_M))/var_M - 1
nu_F = (mu_F*(1 - mu_F))/var_F - 1

alpha_M, beta_M = mu_M*nu_M, (1 - mu_M)*nu_M
alpha_F, beta_F = mu_F*nu_F, (1 - mu_F)*nu_F

x_M = np.linspace(beta.ppf(0.01,alpha_M,beta_M),beta.ppf(0.99,alpha_M,beta_M), 100)
x_F = np.linspace(beta.ppf(0.01,alpha_F,beta_F),beta.ppf(0.99,alpha_F,beta_F), 100)

fig = plt.figure()
ax = plt.axes()

ax.plot(x_M, beta.pdf(x_M,alpha_M,beta_M))
ax.plot(x_F, beta.pdf(x_F,alpha_F,beta_F))

alpha_M, beta_M, alpha_F, beta_F
```
